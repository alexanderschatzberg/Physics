```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#Set up distributions 
sz = 1000000

#Distance to the galactic center ()
d_mean = 8.178
d_std = 0.035
d = np.random.normal(d_mean, d_std, sz)
#convert to meters
d *= 3.086e19

#a. Semi major axis (arcseconds)
a_mean = 0.126
a_std = 0.0011
a = np.random.normal(a_mean, a_std, sz)
a /= 3600 #convert to degrees
#convert to meters
a *= np.pi/180
a *= d


#b. Eccentricity
e_mean = 0.8842
e_std = 0.0023
e = np.random.normal(e_mean, e_std, sz)

#Mass of the black hole (solar masses)
mass_sun = 1.989e30
M_mean = 4.3e6 * mass_sun
M_std = 0.6e6 * mass_sun
M = np.random.normal(M_mean, M_std, sz)

#Speed of light 
c = 299792458

#Gravitational constant
G = 6.67408e-11

```

```python
#Calculate delta phi, the change in orbital phase
delta_phi = (6*np.pi*G*M)/(c**2*a*(1-(e**2)))

#Convert to arcminutes
delta_phi *= 180/np.pi * 60

#Plot the distribution
plt.hist(delta_phi, bins=300) #convert to arcminutes
plt.xlabel('Periastron Advance (arcminutes)')
plt.ylabel('Frequency')
plt.title('Periastron Advance of S2')
```
