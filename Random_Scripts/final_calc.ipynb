```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#Set up distributions for the calculations. The values below are taken directly from the paper
#https://arxiv.org/pdf/2305.10242.pdf (Page 4)

#Size of our normal distributions
sz = 10000000

#Distance to the galactic center (parsec, then meters)
d_mean = 8.178
d_std = 0.035
d = np.random.normal(d_mean, d_std, sz)
#convert to meters
d *= 3.086e19

#a. Semi major axis (arcseconds, converted to meters using the distance to the galactic center)
a_mean = 0.126
a_std = 0.0011
a = np.random.normal(a_mean, a_std, sz)
a /= 3600 #convert to degrees
#convert to meters
a *= np.pi/180
a *= d

#b. Eccentricity of S2's orbit 
e_mean = 0.8842
e_std = 0.0023
e = np.random.normal(e_mean, e_std, sz)

#Mass of the black hole (solar masses converted to kg)
mass_sun = 1.989e30
M_mean = 4.3e6 * mass_sun
M_std = 0.6e6 * mass_sun
M = np.random.normal(M_mean, M_std, sz)

#Speed of light (m/s)
c = 299792458

#Gravitational constant (m^3 kg^-1 s^-2)
G = 6.67408e-11

```

```python
#Calculate delta phi, the change in orbital phase
#This equation is taken from the article I read (https://arxiv.org/pdf/2305.10242.pdf, Page 3)
#It relates eccentricity, the length of the semi-major axis, and the mass of the black hole to periastron advance per orbit 
#delta_phi = (6*np.pi*G*M)/(c^2*a*(1-(e^2)))

delta_phi = (6*np.pi*G*M)/(c**2*a*(1-(e**2)))

#Convert to arcminutes
delta_phi *= 180/np.pi * 60

#Plot the distribution
plt.hist(delta_phi, bins=300) #convert to arcminutes
plt.xlabel('Periastron Advance (arcminutes)')
plt.ylabel('Frequency')
plt.title('Periastron Advance of S2')

#Calculate the mean and standard deviation
mean = np.mean(delta_phi)
std = np.std(delta_phi)

print('Mean (arcminutes): ', mean)
print('Standard Deviation (arcminutes): ', std)
```

```python
#Graph Mass of Black Hole 

#Plot the distribution
plt.hist(M, bins=300)
plt.xlabel('Mass of Black Hole (kg)')
plt.ylabel('Frequency')
plt.title('Mass of Sagittarius A*')

#Calculate the mean and standard deviation
mean = np.mean(M)
std = np.std(M)

print('Mean (kg): ', mean)
print('Standard Deviation (kg): ', std)
```
