```python
import numpy as np
import matplotlib.pyplot as plt

# Define quantities

# Eye height of spearfisher
h = 1.5

# depth of fish
d = 1.0

# index of refraction of medium 1
n1 = 1.0003

# index of refraction of medium 1
n2 = 1.333

def dtheta(t0,h=h,d=d,n1=n1,n2=n2,ret_phi=False):

    # Convert input into radians for calculations
    t = np.radians(t0)
    
    # Actual distance to fish
    x = h*np.tan(t)+d*np.tan(np.arcsin(n1/n2*np.sin(t)))

    # angle 
    phi = np.arctan(x/(h+d))
    if ret_phi:
        return phi
    else:
        dt = np.degrees(t - phi)
        return dt


# Apparant angle of fish
theta = np.linspace(0.0001,89.9999,10000)

# difference in real angle from apparant angle
dt = dtheta(theta)

# Make a plot over a range of 
plt.ion()
plt.figure(1)
plt.clf()

plt.plot(theta, dt,'k-')
plt.xlabel('Apparant angle (degrees)')
plt.ylabel(r'$\Delta\theta$ (degrees)')
plt.title('Spearfishing Problem')
```
