```python
#Libraries 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
#Import datasets
brightest = pd.read_csv("../Data/25BrightestStars.csv")
nearest = pd.read_csv("../Data/25NearestStars.csv")
```

```python
#Define function to generate HR diagram from df 

def hr_diagram(df):
  plt.title('HR Diagram of 25 Nearest and Brightest Stars')

  plt.xlabel("Temperature (K)")
  plt.ylabel("Luminosity (L/Lsun)")

  plt.scatter(df["Temperature"], df["Luminosity"], marker="*")

  plt.xscale("log")
  plt.yscale("log")

  
```

```python
hr_diagram(brightest)
hr_diagram(nearest)

plt.legend(["25 Brightest Stars", "25 Nearest Stars"])

plt.gca().invert_xaxis()
```
