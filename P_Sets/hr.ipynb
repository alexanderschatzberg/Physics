```python
#Libraries 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
#Import datasets
brightest = pd.read_csv("../Data/25BrightestStars.csv")
nearest = pd.read_csv("../Data/25NearestStars.csv")
```

```python
#2, Graph HR Diagram of Nearest and Brightest Stars

def hr_diagram(df):
  plt.title('HR Diagram of 25 Nearest and Brightest Stars')

  plt.xlabel("Temperature (K)")
  plt.ylabel("Luminosity (L/Lsun)")

  plt.scatter(df["Temperature"], df["Luminosity"], marker="*")

  plt.xscale("log")
  plt.yscale("log")

hr_diagram(brightest)
hr_diagram(nearest)

plt.legend(["25 Brightest Stars", "25 Nearest Stars"])

plt.gca().invert_xaxis()
  
```

3. What This Diagram Tells Us 

This diagram shows us that the majority of stars we see in the sky, or at least the brightest ones, are not infact the closest. It also shows us that we have a boatload more small, relativley dim stars close by. While we may see more massive stars in the sky, it's not because they are more common. In fact, the distribution of closer stars more accuratley represents the true distribution of stars in the galaxy. This is because we are measuring the brightness of stars within a region of space, as opposed to looking at them as a function of brightness and distance. 

```python
#4. Plot the HR diagram with lines of constant radius
# L = 4*pi*r^2*sigma*T^4

def hr_diagram(df):
  plt.title('HR Diagram of 25 Nearest and Brightest Stars')

  plt.xlabel("Temperature (K)")
  plt.ylabel("Luminosity (L/Lsun)")

  plt.scatter(df["Temperature"], df["Luminosity"], marker="*")

  plt.xscale("log")
  plt.yscale("log")

hr_diagram(brightest)
hr_diagram(nearest)

plt.legend(["25 Brightest Stars", "25 Nearest Stars"])

plt.gca().invert_xaxis()


boltzmann = 5.670374419e-8 #W/m^2K^4

def constant_radius(radius, temp):
  return 4*np.pi*radius**2*boltzmann*temp**4

temp = np.linspace(start = 1000, stop = 100000, num = 10000)

radius = [0.1, 0.5, 1, 5, 10, 30]

for i in radius:
  plt.plot(temp, constant_radius(i, temp), color="black")
```
