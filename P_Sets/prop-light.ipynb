```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Problem 3
#Delta theta for every degree between 30 and 90

n1 = 1.0
n2 = 1.0003

altitudes = np.arange(30,90,0.1)

theta = np.arcsin(n1*np.cos(np.radians(altitudes))/n2)

d_theta = (90 - altitudes) - np.degrees(theta)

#Plotting
plt.plot(altitudes, d_theta)
plt.xlabel('Angle of altitude (degrees)')
plt.ylabel('Change in angle (degrees)')
```
