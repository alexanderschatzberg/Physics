```python
#Imports
import numpy as np
import matplotlib.pyplot as plt 
```

```python
#Pendulum Measurements
time = np.array([0, 20.3, 35.8, 39.9, 42.9, 77.8, 92.1, 120.7, 153.3, 205.4, 218.4, 242.1, 334.9, 447.9])
time_sigma = 0.1

amp_decay = np.array([30, 28, 27, 26, 25, 22, 20, 18, 15, 12, 11, 10, 6, 4])
amp_sigma = 2
```

```python
#Plotting the data with error bars for the pendulum
"""

"""
plt.figure(1)
plt.plot(time, amp_decay, 'o', color='Black') #Plot the points

plt.errorbar(time, amp_decay, xerr=time_sigma, yerr=amp_sigma, fmt='o', ecolor='Black') #Plot the error bars

#Label the axises and title the graph
plt.title('Amplitude Decay of a Pendulum')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude (deg)')

#Fit a line to the data (Ae^(-x/tau))
tau = 205 #Seconds. Tau = the time it takes for the amplitude to decay to Amp nought/e
tau_err = 1

y = 30*np.exp(-time/tau)
plt.plot(time, y, color='Red')  

#Show the plot
plt.show()
```
